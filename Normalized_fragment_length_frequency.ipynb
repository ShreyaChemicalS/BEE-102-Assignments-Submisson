```python
import pandas as pd
```

```python
data = pd.read_csv('/content/query.bed', sep="\t", header=None)
```

```python
data.head()
```

```python

data.columns = [ "Chromosomes", "Left Position", "Right Position", 'c2', 'c3', 'ChrLengths']  # or whatever you want!

print(data)
```

```python
data['Chrlength'] = data['Right Position'] - data['Left Position']
```

```python
data.head()
```

```python
Sum = data['Chrlength'].sum()
```

```python
Sum
```

```python
length_counts = data["Chrlength"].value_counts().sort_index()

normalized = length_counts / length_counts.sum()
```

```python
normalized
```

```python
import matplotlib.pyplot as plt
```

```python
plt.plot(normalized.index, normalized.values)
plt.xlabel('Fragment Length')
plt.ylabel('Normalized Frequency [A.U.]')
plt.title('Normalized Chromosome Length Frequency')
plt.show()
```
